```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.arange(1,11)
y = 2 * x + 5
plt.title("Matplotlib demo")
plt.xlabel("x axis caption")
plt.ylabel("y axis caption")
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(1,11)
y = 2 * x + 5
plt.title("Matplotlib demo")
plt.xlabel("x axis caption")
plt.ylabel("y axis caption")
plt.plot(x,y,"ob")
plt.show()
```

### Sine Wave Plot

```python
x = np.arange(0,3 * np.pi,0.1)
y = np.sin(x)
plt.title("sine wave form")
plt.plot(x,y)
plt.show()
```

### subplot()

```python
x = np.arange(0,3 * np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.subplot(2,1,1)
plt.plot(x,y_sin)
plt.title("sine")
plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.show()
```

### bar()

```python
x = [5,8,10]
y = [12,16,6]

x2 = [6,9,11]
y2 = [6,15,7]

plt.bar(x,y, align= "center")
plt.bar(x2,y2,color = "g" , align= "center")
plt.title("Bar Graph")
plt.ylabel("Y axis")
plt.xlabel("X axis")
```
