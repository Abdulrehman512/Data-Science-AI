```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("tested.csv")
```

### Count Plot

```python
sns.countplot(df["Survived"])
```

### Pie Chart

```python
df["Survived"].value_counts().plot(kind="pie", autopct = "%.2f")
```

### Histogram

```python
plt.hist(df["Age"])
```

### Dist Plot

```python
sns.displot(df['Age'])
```

### Box Plot

```python
sns.boxplot(df['Fare'])
```
