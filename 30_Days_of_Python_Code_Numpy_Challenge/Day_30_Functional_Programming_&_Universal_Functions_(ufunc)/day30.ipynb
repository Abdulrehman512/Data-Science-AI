### Q1:

```python
import numpy as np
 
 
A = np.array(
    [
        [4, 3, 6],
        [5, 2, 7],
        [8, 3, 10],
        [-3, 4, 2],
    ]
)
 
 
def calc_diff(row):
    return max(row) - min(row)
 
 
result = np.apply_along_axis(calc_diff, 1, A)
print(result)
```

### Q2:

```python
import numpy as np
 
 
A = np.array(
    [
        [4, 3, 6],
        [5, 2, 7],
        [8, 3, 10],
        [3, 4, 2],
    ]
)
 
 
def standardize(column):
    max_val = max(column)
    min_val = min(column)
    return (column - min_val) / (max_val - min_val)
 
 
result = np.apply_along_axis(standardize, 0, A)
print(result)
```

### Q3:

```python
import numpy as np
 
 
A = np.array(
    [
        [4, 3, 6],
        [5, 2, 7],
        [8, 3, 10],
        [-3, 4, 2],
    ]
)
 
 
def swap_values(row):
    first = row[0]
    last = row[-1]
    row[0] = last
    row[-1] = first
    return row
 
 
result = np.apply_along_axis(swap_values, 1, A)
print(result)
```

### Q4:

```python
import numpy as np
 
 
A = np.array([4, 5, 1, -4])
B = np.array([[3, 6, 7, 9], [-1, 2, 5, 4]])
 
 
def add_three(a):
    return a + 3
 
 
add_three = np.frompyfunc(add_three, 1, 1)
 
print(add_three(A))
print(add_three(B))
```

### Q5:

```python
import numpy as np
 
 
A = np.array([4, 5, 1, -4])
B = np.array([3, -6, 7, 9])
 
 
def add_abs(a, b):
    return abs(a) + abs(b)
 
 
add_abs = np.frompyfunc(add_abs, 2, 1)
 
print(add_abs(A, B))
```
