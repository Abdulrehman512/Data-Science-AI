### Q1:

```python
import numpy as np
 
 
v = np.array([3, 4, -2])
 
v_norm = np.linalg.norm(v)
print(v_norm)
```

### Q2:

```python
import numpy as np
 
 
A = np.array([[2, 3], [-4, 2], [5, 0]])
B = np.array([[4, 3, 2], [-1, 0, 2]])
 
result = np.dot(A, B)
print(result)
```

### Q3:

```python
import numpy as np
 
 
arr = np.array(
    [
        [-2, 0, 4],
        [5, 2, -1],
        [-4, 2, 4],
    ]
)
 
arr_det = np.linalg.det(arr)
print(arr_det)
```

### Q4:

```python
import numpy as np
 
 
arr = np.array(
    [
        [5, 8, 16],
        [4, 1, 8],
        [-4, 4, -11],
    ]
)
 
eigenvalues, eigenvectors = np.linalg.eig(arr)
 
print(eigenvalues)
print(eigenvectors)
```
