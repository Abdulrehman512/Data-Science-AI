### Q1:

```python
import numpy as np
 
 
np.random.seed(42)
arr = np.random.randint(low=0, high=2, size=(10, 6))
 
result = np.count_nonzero(arr)
print(result)
```

### Q2:

```python
import numpy as np
 
 
np.random.seed(42)
arr = np.random.randn(10, 4)
np.set_printoptions(precision=4)
print(arr)
```

### Q3:

```python
import numpy as np
 
 
arr = np.array([1.2e-6, 1.7e-7])
np.set_printoptions(suppress=True, precision=8)
print(arr)
```
