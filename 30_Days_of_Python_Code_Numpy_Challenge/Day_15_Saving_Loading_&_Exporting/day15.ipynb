### Q2:

```python
import numpy as np
 
 
arr = np.arange(12, dtype='float32').reshape(-1, 4)
np.savetxt(fname='array.txt', X=arr, fmt='%0.2f')
 
loaded_arr = np.loadtxt('array.txt')
print(loaded_arr)
```

### Q4:

```python
import numpy as np
 
 
arr = np.array(
    [
        [
            [0.0, 2.0, 0.5, 1.0],
            [0.0, 2.0, 0.0, 3.0],
            [0.0, 4.0, 0.0, 0.0],
        ],
        [
            [0.0, 0.5, 0.0, 0.2],
            [0.0, 0.4, 0.0, 5.0],
            [0.0, 0.0, 1.0, 0.0],
        ],
    ]
)
 
np.savetxt('arr1.csv', arr[0, :, :], delimiter=',')
np.savetxt('arr2.csv', arr[1, :, :], delimiter=',')
 
arr1 = np.loadtxt('arr1.csv', delimiter=',')
arr2 = np.loadtxt('arr2.csv', delimiter=',')
 
result = np.c_[[arr1.T, arr2.T]]
 
print(result)
```

```python
import numpy as np
 
 
with open('stocks.txt', 'r') as file:
    text = file.read()
 
lines = text.splitlines()
lines = [line.split('\t') for line in lines]
result = np.array(lines, dtype=str)
 
print(result)
```
