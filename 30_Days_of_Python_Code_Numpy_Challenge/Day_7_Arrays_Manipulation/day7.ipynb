### Q2:

```python
import numpy as np
 
 
np.random.seed(42)
 
image = np.random.randint(
    low=0, 
    high=256, 
    size=(10, 10, 3), 
    dtype=np.uint8
)
 
print(image[:, :, 0])
```

### Q4:

```python
import numpy as np
 
 
np.random.seed(42)
 
arr = np.array(
    [
        ['id', 'price'],
        ['001', '14.99'],
        ['002', '4.99'],
        ['003', '7.99'],
        ['004', '2.49'],
        ['005', '1.49'],
    ]
)
 
np.random.shuffle(arr[1:])
print(arr)
```
