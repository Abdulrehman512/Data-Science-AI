### Q5:

```python
import numpy as np
 
 
arr = np.array(
    [
        [4, 3, 6],
        [5, 2, 7],
        [8, 3, 10],
        [-3, 4, 2],
    ]
)
 
 
def calc_diff(column):
    return max(column) - min(column)    
 
 
diff = np.apply_along_axis(calc_diff, 0, arr)
print(diff)
```
