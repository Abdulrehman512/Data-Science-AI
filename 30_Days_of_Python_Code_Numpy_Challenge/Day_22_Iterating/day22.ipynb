## Q5:

```python
import numpy as np
 
 
arr = np.zeros((10, 4), dtype='int')
 
for (i, j), _ in np.ndenumerate(arr):
        if (i + j) % 3 == 0:
            arr[i, j] = 1
 
print(arr)
```

## Q6:

```python
import numpy as np
 
 
arr = np.zeros((9, 9), dtype='int')
 
for (i, j), _ in np.ndenumerate(arr):
    arr[i, j] = i ** 2 + j ** 2
 
arr_sqrt = np.sqrt(arr)
 
for (i, j), val in np.ndenumerate(arr_sqrt):
    arr[i, j] = int(val) == val
 
print(arr)
```
