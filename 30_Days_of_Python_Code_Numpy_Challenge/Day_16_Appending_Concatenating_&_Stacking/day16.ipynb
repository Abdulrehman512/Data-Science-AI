```python
import numpy as np
 
 
arr1 = np.arange(12).reshape(-1, 4)
arr2 = np.array([[4, 3, 7, 2], [0, 5, 2, 6]])
result = np.append(arr1, arr2, axis=0)
 
print(result)
```

### Q2:

```python
import numpy as np
 
 
arr1 = np.array([[3, 4, 5], [8, 3, 1]])
arr2 = np.array([[0, 5, 2], [4, 2, 1]])
result = np.concatenate((arr1, arr2), axis=0)
 
print(result)
```

### Q6:

```python
import numpy as np
 
 
def shift(arr, n):
    if n == 0:
        return arr
    elif n >= 0:
        return np.r_[np.full(n, np.nan, dtype=arr.dtype), arr[:-n]]
    else:
        return np.r_[arr[-n:], np.full(-n, np.nan, dtype=arr.dtype)]
```
