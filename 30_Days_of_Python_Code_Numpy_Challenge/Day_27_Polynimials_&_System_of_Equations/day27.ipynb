### Q3:

```python
import numpy as np
 
 
W = np.array([4, 5, 1])
Q = np.array([2, 4, -5, 1])
 
WQ_add = np.polyadd(W, Q)
WQ_sub = np.polysub(W, Q)
WQ_mul = np.polymul(W, Q)
W2Q_add = np.polyadd(W, 2 * Q)
 
print(WQ_add)
print(WQ_sub)
print(WQ_mul)
print(W2Q_add)
```

### Q4:

```python
import numpy as np
 
 
A = np.array([[5, -3], [1, -2]])
B = np.array([[21], [7]])
 
X = np.dot(np.linalg.inv(A), B)
x = X[0][0]
y = X[1][0]
 
print(f'x = {x:.2f}\ny = {y:.2f}')
```

### Q5:

```python
import numpy as np
 
 
A = np.array([[1, 1, 1], [2, 1, 5], [1, -1, -1]])
B = np.array([[1], [0], [0]])
 
X = np.dot(np.linalg.inv(A), B)
x = X[0][0]
y = X[1][0]
z = X[2][0]
 
print(f'x = {x:.2f}\ny = {y:.2f}\nz = {z:.2f}')
```
