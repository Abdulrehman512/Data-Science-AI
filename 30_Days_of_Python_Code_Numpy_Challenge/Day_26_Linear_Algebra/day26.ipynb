### Q4:

```python
import numpy as np
 
 
v1 = np.array([1, 0, 0])
v2 = np.array([0, 1, 0])
 
v1_norm = np.linalg.norm(v1)
v2_norm = np.linalg.norm(v2)
 
result = np.arccos(np.dot(v1, v2) / (v1_norm * v2_norm))
print(result)
```

### Q5:

```python
import numpy as np
 
 
v1 = np.array([1, 0, 0])
v2 = np.array([0, 1, 0])
 
v1_norm = np.linalg.norm(v1)
v2_norm = np.linalg.norm(v2)
 
result = np.degrees(np.arccos(np.dot(v1, v2) / (v1_norm * v2_norm)))
print(result)
```

### Q6:

```python
import numpy as np
 
 
u = np.array([1, 0, 0])
v = np.array([0, 1, 0])
 
is_orthogonal = np.dot(u + v, u + v) == np.dot(u, u) + np.dot(v, v)
 
print(is_orthogonal)
```
